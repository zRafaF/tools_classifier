```python
import os
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
source_dir = './tools_classifier/data_set/ground_truth_raw'
target_dir = "binarized_dataset"
```

```python
def binarize_and_show(image_path, output_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    _, binary_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    cv2.imwrite(output_path, binary_img)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1)
    plt.imshow(img)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.imshow(binary_img, cmap='gray')
    plt.title("Binarized Image")
    plt.axis('off')

    plt.show()

```

```python
def process_dataset(source, target):
    for root, dirs, files in os.walk(source):
        # Get relative path to preserve structure
        relative_path = os.path.relpath(root, source)
        target_folder = os.path.join(target, relative_path)

        if not os.path.exists(target_folder):
            os.makedirs(target_folder)

        # Process each file
        for file in files:
            if file.endswith(".png"):
                image_path = os.path.join(root, file)
                output_path = os.path.join(target_folder, file)
                binarize_and_show(image_path, output_path)

process_dataset(source_dir, target_dir)
```
